```python
def inputData(stats):
    counter = 0;
    for item in stats:
        a = pd.read_csv(item)
        a["TMYR"] = a.apply(lambda x: "{0}{1}".format(a["Tm"], counter), axis = 1)
        counter++;
        data[item.replace(".csv", "")] = a
```

```python
def combineAndFilter(var):
    df = inputData(var)
    df = df[df['Tm'] != 'TOT']
    g = df.shape[0]
    for i in range(0, g):
        if i < g -1 and (df.iloc[i, 1] == df.iloc[i+1, 1]):
            if(df.iloc[i, 5] >= 30):
                df.iat[i, 5] = 50
            elif(df.iloc[i+1, 5] >= 30):
                df.iat[i+1, 5] = 50
    df = df[df['G'] >= 50]
    df["PS/G\\"] = [s.replace('\\', '') for s in df["PS/G\\"]]
    df["PS/G\\"] = [s.replace('nan', '') for s in df["PS/G\\"]]
    df["PS/G\\"] = [s for s in df["PS/G\\"] if s!= '']
    df["PS/G\\"] = df["PS/G\\"].astype(float)
    p = np.percentile(df["PS/G\\"], 75)
    print(p)
    df = df[df["PS/G\\"] >= p]
    return df
```
